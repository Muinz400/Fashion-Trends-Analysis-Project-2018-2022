```python
import pandas as pd

# Load the exported CSV
df = pd.read_csv("/Users/manyangtaal/Downloads/CPlusPlus.cpp/sql_project/exports/q1_seasonal_fashion_trends.csv")

# Take a peek at the data
df.head(40)
```

```python
# Group by category + season and calculate total sales
seasonal_trends = df.groupby(["category", "season"])["total_sales"].sum().reset_index()

# Sort descending, same as ORDER BY total_sales DESC
seasonal_trends = seasonal_trends.sort_values(by="total_sales", ascending=False)

seasonal_trends.head(40)  # show top 10
```
